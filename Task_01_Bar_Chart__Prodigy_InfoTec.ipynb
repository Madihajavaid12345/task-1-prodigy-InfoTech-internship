```python
# Step 1: Import necessary libraries
import pandas as pd              # For working with data
import matplotlib.pyplot as plt  # For creating visualizations

# Optional: Use a dark background style
plt.style.use('dark_background')
```

```python
# Step 3: Check available years and choose the most recent populated column (e.g., '2024')
years = [col for col in df.columns if col.isdigit()]
target_year = years[-1]  # Most recent year, e.g., '2024'
```

```python
# Step 2: Load the main population CSV (skip top metadata rows)
df = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv", skiprows=4)
```

```python


# Step 4: Prepare data for plotting
data = df[['Country Name', target_year]].dropna()
```

```python





# Bar Chart ‚Äî Top 10 Countries
top10 = data.sort_values(by=target_year, ascending=False).head(10)
plt.figure(figsize=(12, 6))
plt.bar(top10['Country Name'], top10[target_year]/1e6, color='deepskyblue')
plt.title(f"üåç Top 10 Most Populated Countries ‚Äì {target_year}", color='cyan', fontsize=16)
plt.xlabel("Country", fontsize=12)
plt.ylabel("Population (Millions)", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


```

```python

# Histogram ‚Äî Distribution of Populations Across All Countries
pop_millions = data[target_year] / 1e6
plt.figure(figsize=(10, 5))
plt.hist(pop_millions, bins=20, color='magenta', edgecolor='white', alpha=0.8)
plt.title(f"üìà Distribution of Country Populations ‚Äì {target_year}", color='cyan', fontsize=16)
plt.xlabel("Population (Millions)", fontsize=12)
plt.ylabel("Number of Countries", fontsize=12)
plt.tight_layout()
plt.show()
```
